```python
import pandas as pd
from pathlib import Path

RAW_DIR = Path("/Users/bharathreddy97/Documents/RevSense-AI/data/raw")

for file in RAW_DIR.glob("*.csv"):

    df = pd.read_csv(file)

    print("=" * 80)
    print(f"TABLE: {file.name}")
    print("=" * 80)

    # print(df.head)

    print("Shape:")
    print(df.shape)

    print("\nColumns:")
    print(df.columns.tolist())

    print("\nData Types:")
    print(df.dtypes)

    print("\nTop Null Columns:")
    print(df.isnull().sum().sort_values(ascending=False).head())

    print("\nPotential Unique Columns:")
    for col in df.columns:
        if df[col].nunique() == len(df):
            print(col)

    print("\n")
```
